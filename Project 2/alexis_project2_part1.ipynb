```python
df_new=df.copy()
```

```python
threshold = 50
missing_percent=(df_new.isnull().sum()/len(df_new))*100
cols_to_drop=missing_percent[missing_percent > threshold].index
df_new=df_new.drop(columns=cols_to_drop)
df_new
```

```python
df_new.isnull().sum()[df_new.isnull().sum() > 0] 
```

```python
print(df_new[df_new['Engine Cylinders'].isnull()]['Fuel Type'].unique())
```

```python
df_new.loc[df_new['Fuel Type'] == 'Electricity', 'Engine Cylinders'] = 0
df_new.loc[df_new['Fuel Type'] == 'Electricity', 'Engine Displacement'] = 0
df_new
```

```python
#because these columns are numerical, i can input the median for the missing values of non-electric vehicles and this will keep the distribution stable
df_new['Engine Cylinders']=df_new['Engine Cylinders'].fillna(df['Engine Cylinders'].median())
df_new['Engine Displacement']=df_new['Engine Displacement'].fillna(df['Engine Displacement'].median())
df_new
```

```python
#i'm going to label the missing categorical values with "unknown"
df_new['Drive']=df_new['Drive'].fillna('Unknown')
df_new['Transmission']=df_new['Transmission'].fillna('Unknown')
df_new['Engine Descriptor']=df_new['Engine Descriptor'].fillna('Unknown')
df_new
```

I referenced the notebook that you provided us with to better clean the data. The columns that were dropped would've interfered with my analysis because they had a large percent of missing values that would have been difficult to work with. The columns that I filled with the median and 'unknown' were also for ease of analysis. It would've been harder to use the dataset otherwise.

```python
df_new.duplicated().sum()
```

```python
#i'm going to add a column of cost per mile, i think this will help identify the most cost-efficient cars
df_new['Cost Per Mile']=df_new['Annual Fuel Cost (FT1)']/15000
print(df_new[['Make', 'Model','Cost Per Mile']].head())
```

```python
#curious about seeing the high efficiency vehicles in 2017
df_filtered=df_new[(df_new['Year'] == 2017)& (df_new['Combined MPG (FT1)'] > 30)]
print(df_filtered[['Make', 'Model', 'Year', 'Combined MPG (FT1)']])
```

```python
#unique values of a categorical column
print(df_new['Drive'].unique())
```

Because we're interested in seeing the most efficient vehicle, categorical columns like Make, Model, Drive, Fuel Type would be important for our analysis

```python
make_grouping=df_new.groupby('Make')['Combined MPG (FT1)'].mean()
make_grouping=make_grouping.sort_values(ascending=False)
detailed=df_new.groupby('Make')['Combined MPG (FT1)'].agg(['mean', 'count', 'max'])
print(detailed.sort_values(by='mean', ascending=False).head(10))
```

This grouping helped me extract more specific data from the dataset. By grouping by Make, I was able to see that electric/hybrid manufacturers usually display the highest average fuel economies.

```python
#the car with the highest MPG
best_car=df_new.loc[df_new['Combined MPG (FT1)'].idxmax()]
print(f'The most efficient vehicle is the {best_car['Year']} {best_car['Make']} {best_car['Model']}.')
```
