### Importing Libraries

```python
import numpy as np
import pandas as pd
```

```python
n = 10 
np.random.seed(42) 
first_name=['Alexis', 'Tira', 'Amiyah', 'Kenny', 'Makayla', 'Star', 'Tre', 'Ray', 'Yeni', 'Maddie']
last_name=['Smith', 'Johnson', 'Brown', 'Taylor', 'Anderson']
name=np.random.choice(first_name,n)+" "+np.random.choice(last_name,n) 
print(name)
ages=np.random.randint(18,26,n)
marks=np.random.randint(40,100,size=(n,5))
gender=np.random.choice(['Male', 'Female'],size=n)
print(ages)
print(marks)
student_data=pd.DataFrame(marks,columns=['Physics', 'Math', 'CS', 'English', 'History'])
student_data
```

```python
student_data.insert(0,'Age',ages)
```

```python
student_data.insert(0,'Name',name)
```

```python
student_data.insert(2,'Gender',gender)
```

```python
student_data
```

```python
student_data.to_csv('student_raw_data.csv',index=False)
```
