# Indexing

```python
import numpy as np
import pandas as pd
```

```python
df_new=pd.read_csv('student_raw_data_new.csv')
df_new
```

* loc (label-based)
* iloc (number-based)

```python
df_new.iloc[0]
```

```python
df_new.loc[0]
```

```python
print(df_new.iloc[0,4])
```

```python
print(df_new.loc[0,'Math'])
```

```python
df_new.iloc[1:3]
```

```python
df_new.loc[1:3]
```

```python
df_index=df_new.set_index('Students_Name')
df_index
```

```python
df_index.iloc[0]
```

```python
df_index.loc['Tre Taylor']
```

```python
df_sorted=df_new.sort_values(by='Physics',ascending=False)
df_sorted
```

## Group by

```python
df_new.groupby('Gender')[['Physics', 'Math', 'CS']].mean()
```

```python
df_new.groupby('Gender')[['Physics', 'Math', 'CS']].agg(['mean', 'max', 'count'])
```
