```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('student_raw_data.csv')
df
```

```python
df_new=df.copy()
df_new
```

```python
df_new['Age']
```

```python
#df_new['Total']=df_new['Physics']+df_new['Math']+df_new['CS']+df_new['English']+df_new['History']
df_new['Total']=df_new[['Physics', 'Math', 'CS', 'English', 'History']].sum(axis=1) #axis=1 is embedding a new column, axis=0 represents a row
```

```python
df_new
```

```python
df_new['Sci_total']=df_new[['Physics', 'CS']].sum(axis=1)
```

```python
df_new
```

```python
df_new['Percentage']=(df_new['Total']/500)*100
df_new
```

```python
def grade(Percentage):
    if Percentage >=90:
        return 'A'
    elif Percentage >=80:
        return 'B'
    elif Percentage >=70:
        return 'C'
    elif Percentage >=60:
        return 'D'
    else:
        return 'F'
```

```python
df_new['Grade']=df_new['Percentage'].apply(grade)
df_new
```

```python
df_new['Percentage_bonus']=df_new['Percentage'].apply(lambda x: x+5)
df_new
```

```python
print('Data Frame before dropping bonus column:\n', df_new)
df_new.drop(columns=['Percentage_bonus'], inplace=True)
print('Data Frame after dropping bonus column:\n', df_new)
```

```python
df_new
```

```python
df_filtered=df_new[df_new['Physics']>70]
df_filtered
```

```python
df_A=df_new[df_new['Percentage']>=90]
df_A
```

```python
filtered_df=df_new[(df_new['Physics']>60)&(df_new['Math']>60)]
filtered_df
```

```python
df_new.rename(columns={'Name':'Students_Name'},inplace=True)
df_new
```

```python
df_index=df_new.set_index('Students_Name')
df_index
```

```python
df_new.to_csv('student_raw_data_new.csv', index=False)
```
