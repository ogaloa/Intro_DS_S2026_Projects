```python
import numpy as np
import pandas as pd
```

```python
n=200
np.random.seed(42)
user=['Alexis','Danya','James','Kendra','Elliot','Amiyah','Dante','Rose','Ben','Elaine']
device_model=['Iphone13','Samsung','Iphone15','Google_pixel','Iphone16']
data={
    'User':np.random.choice(user,size=n),
    'Device_Model':np.random.choice(device_model,size=n),
    'Screentime(hrs)':np.random.randint(1,12,size=n)
}
ages=np.random.normal(loc=30, scale=8, size=n).round(1)
gender=np.random.choice(['Male','Female'],size=n)
mobile_data=pd.DataFrame(data)
mobile_data
```

```python
mobile_data.insert(1,'Gender',gender)
```

```python
mobile_data.insert(2,'Age',ages)
```

```python
mobile_data
```

```python
mobile_data.to_csv('mobile_data_raw.csv',index=False)
```
